```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#### checking for imbalance in the dataset
df.quality.value_counts().plot(kind='bar', color='pink',edgecolor='red')
plt.xlabel("Quality of Wine")
plt.ylabel("Count")
plt.show()
```

data about wines with quality=7, quality=8, quality=4, and quality=10 are much less than the majority data therefore it can be conculded that this is an imbalanced dataset

```python
## Check distribution of each column in the dataset
for column in df.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[column], kde=True)
    plt.title(f"Distribution of {column}")
    plt.show
```

Univariate exploration:

```python
## 1. for alcohol
sns.histplot(df['alcohol'], kde=True)
sns.boxplot(x=df['alcohol'])
```

```python
## 2. for ph
sns.histplot(df['pH'], kde=True)
sns.boxplot(x=df['pH'])
```

```python
## 3. for density
sns.histplot(df['density'], kde=True)
sns.boxplot(x=df['density'])
```

Bivariate exploration:


```python
#### Alcohol vs density
sns.scatterplot(x='alcohol', y='density', data=df)
```

```python
#### Bivariate plot for categorical attributes
sns.catplot(x='quality',y='alcohol',data=df,kind='box')
```

```python
# Alcohol vs ph by wine quality

sns.scatterplot(x='alcohol',y='pH',hue='quality',data=df)
```

Multivariate Analysis:

```python
### Pairplot for key numeric variables.
sns.pairplot(df[['alcohol', 'pH', 'density', 'quality']])
```
